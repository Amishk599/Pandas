```python
import pandas as pd
df=pd.read_csv(r"C:\Users\amish\Desktop\weather_data_2.csv")
df
```

```python
g = df.groupby("City")# This statement creates an object that has multiple data frames grouped by respective cities
g # DataFrameGroupBy object
```

# To the display the entire grouped data frame 

```python
for City, df_City in g:
    print(City)
    print(df_City)
```

# Displaying the specific grouped data frame- get_group()

```python
g.get_group("Mumbai")
```

# Split, Apply and Combine in grouped data 

```python
g.max() # This function diplays the max value of each column of each grouped DataFrame
```

```python
g.mean()
```
