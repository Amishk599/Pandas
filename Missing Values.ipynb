```python
import pandas as pd
df = pd.read_csv(r"C:\Users\amish\Desktop\weather_data.csv", parse_dates=["Date"])
df
type(df.Date[0]) # elements under Dates now being treated as timestamps
```

```python
df.set_index("Date",inplace=True) #inplace=True edits the original data frame 
df
```

# Using df.fillna() 

```python
ef = df.fillna(0)
ef # fills not available values with passed value
```

```python
ef = df.fillna({
    "Temperature": 0,
    "Windspeed": 0,
    "Event": "No Event"
}) # passing a dictionary to specify the kind of data to be filled in place of missing values
ef
```

```python
ef = df.fillna(method="ffill")# forward fill fills the context of one cell into consecutive missing valued cells, similarly, bfill does the opposite
ef
```

# Guessing appropriate values using interpolation

```python
df
```

```python
ef = df.interpolate() #fills intermediate appropriate value for missing cells considering the values above and below
ef
```

```python
ef = df.interpolate(method="time")# sincle data for 2017-01-03/02 were not available so data for 5th was evaluated by even considering the ones on those dates
ef
```

# Dropping rows with NaN values 

```python
df
```

```python
ef = df.dropna() # drops all the rows with even atleast one NaN
ef
```

```python
ef = df.dropna(how="all")# removes the rows with all the fields as NaN; not including index
ef
```

```python
ef = df.dropna(thresh=1)#thresh leaves all the rows with alteast one valid value; eliminating the rest
ef
```

```python
ef = df.dropna(thresh=2)#thresh leaves all the rows with alteast two valid value; eliminating the rest
ef
```
