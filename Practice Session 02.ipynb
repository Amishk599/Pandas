## Using DataFrame.loc[] 

**Purpose: Access a group of rows and columns by label(s) or a boolean array.**

***.loc[] is primarily label based.***

```python
df.loc[5:10] # numbers are treated as labels of index
```

```python
gf = df.set_index("company")
gf
```

```python
gf.loc["audi"] #accessing all the label specific rows
```

```python
gf.loc["audi","price"] # accessing rows and column specific values
```

**Details of the most expensive car**

```python
df.loc[df["price"] == df["price"].max()]   # details of car with the highest cost
```

**Printing details of all the Toyota cars -** *using .loc[ ] and groupby()*

```python
df.loc[df["company"]=="toyota"] #using .loc[]
```

```python
manufacturer = df.groupby("company") # By creating a groupby() object
manufacturer.get_group("toyota")
```

## concat()

*pandas.concat(objs, axis=0, join='outer', join_axes=None, ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=None, copy=True)*

```python
GermanCars = pd.DataFrame({
    'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 
    'Price': [23845, 171995, 135925 , 71400]
})
JapaneseCars = pd.DataFrame({
    'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '],
    'Price': [29995, 23600, 61500 , 58900]
})

pd.concat([GermanCars,JapaneseCars],keys=["German","Japanese"])
```

```python
car_price = { "Company": ['Toyota', 'Honda', 'BMV', 'Audi'], 
    'Price': [23845, 17995, 135925 , 71400]}
car_horsePower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 
                  'horsepower': [141, 80, 182 , 160]}
car_price_df = pd.DataFrame.from_dict(car_price)
car_horsePower_df = pd.DataFrame.from_dict(car_horsePower)
```

```python
car_price_df
```

```python
car_horsePower_df
```

## Merging the two dataframes

```python
pd.merge(car_price_df, car_horsePower_df, on = "Company")
```
