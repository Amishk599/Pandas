## Reshaping dataframe using melt()

pandas.melt(frame, id_vars=None, value_vars=None, var_name=None, value_name='value', col_level=None)

```python
import pandas as pd
Temperatures={
    "Day": ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"],
    "Chicago": [32,33,31,32,38,39,37],
    "Mumbai": [39,40,38,41,42,38,36],
    "Los Angeles": [23,24,25,21,19,26,23]
}
temperature_df=pd.DataFrame.from_dict(Temperatures)
temperature_df

```

```python
df1=pd.melt(temperature_df,id_vars="Day")
df1
```

```python
df2=pd.melt(temperature_df,id_vars="Day",var_name="City",value_name="Temperature") #column headers defined
df2
```
