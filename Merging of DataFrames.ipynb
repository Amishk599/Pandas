```python
# Creating the dataframes to be merged
```

```python
import pandas as pd
df1 = pd.DataFrame({
    "City": ["Los Anglese","San Francisco","Seattle","Miami","San Diego"],
    "Temperature": [18, 12, 2, 23, 15]
})
df1
```

```python
df2 = pd.DataFrame({
    "City": ["Los Anglese","San Francisco","Seattle","Miami","Las Vegas", "Yosemite"],
    "Humidity": [20, 69, 88, 45, 46, 10]
})
df2
```

# Merging (intersection of) two data frames

```python
df3 = pd.merge(df1, df2, on ="City")# by default merge finds the intersection of the two data frames
df3
```

# To find the union of the two data frames use the "how" option as "outer"

```python
df4 = pd.merge(df1, df2, on="City", how = "outer", indicator=True)# indicator flag shows the orgin of each data
df4
```

```python
df4 = pd.merge(df1, df2, on="City", how = "left")# merging as intersection and left circle
df4
```

```python
df4 = pd.merge(df1, df2, on="City", how = "outer")# merging as intersection and right circle
df4
```
